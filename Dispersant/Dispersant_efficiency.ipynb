# ML models for oil dispersant efficiency prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import cross_val_score,cross_val_predict
#Evaluation metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
#Linear Regression model
from sklearn.linear_model import LinearRegression
#Ridge Regressor model
from sklearn.linear_model import Ridge
#Decision Trees model
from sklearn.tree import DecisionTreeRegressor
#Random forest model
from sklearn.ensemble import RandomForestRegressor
#XGBoost model
from xgboost import XGBRegressor
# Gradient Boost model
from sklearn.ensemble import GradientBoostingClassifier,GradientBoostingRegressor
```

```python
df.columns
```

```python
X=df.drop(columns=['Эффективность %','Месторождение','соотношение'],axis=1)
y=df['Эффективность %']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

# Baseline ML Models

```python
# Baseline Models
dtree=DecisionTreeRegressor(random_state=42)
rf=RandomForestRegressor(random_state=42)
xgb=XGBRegressor(random_state=42)
lr=LinearRegression()
gb=GradientBoostingRegressor(random_state=42)
base_models= [dtree, rf, xgb, lr, gb]
base_model_string=['Decision Trees','Random Forest','XGBoost', 'Linear Regression','Gradient Boosting']
```

```python
# Performance of baseline models
i=0
MAE_base=[]
MSE_base=[]
RMSE_base=[]
R2_base=[]
EVS_base=[]
for model in base_models:
    model.fit(X_train,y_train)
    y_pred_final=model.predict(X_test)
    print("Performance evaluation of {} model".format(base_model_string[i]))
    mae_base=mean_absolute_error(y_test,y_pred_final)
    print("MAE=",mae_base)
    mse_base=mean_squared_error(y_test,y_pred_final)
    print("MSE=",mse_base)
    rmse_base=np.sqrt(mean_squared_error(y_test,y_pred_final))
    print("RMSE=",rmse_base)
    r2_base=r2_score(y_test,y_pred_final)
    print("R^2 score=",r2_base)
    evs_base=explained_variance_score(y_test,y_pred_final)
    print("Explained_variance=",evs_base)
    i+=1
    print("\n")
    MAE_base.append(mae_base)
    MSE_base.append(mse_base)
    RMSE_base.append(rmse_base)
    R2_base.append(r2_base)
    EVS_base.append(evs_base)
    
```

# Model Comparison Visualization of Baseline Models

```python
# Model Comparison of Baseline Models
df_baseline=pd.DataFrame({
    'Baseline ML Model':base_model_string,
    'MAE': MAE_base,
    'MSE': MSE_base,
    'RMSE': RMSE_base,
    'r^2 score':R2_base,
    'Explained Variance Score':EVS_base
})
df_baseline
```

# Hyperparameter Tuning of Baseline ML Models

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameters for each model
dtree_params = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

xgb_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.5, 0.7, 1.0],
    'colsample_bytree': [0.5, 0.7, 1.0],
    'gamma': [0, 0.1, 0.3]
}

lr_params = {
    # No hyperparameters to tune for Linear Regression
}

gb_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.3],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Define parameter grids for each model
param_grids = [dtree_params, rf_params, xgb_params, lr_params, gb_params]

# Perform GridSearchCV for each model
best_estimators = []
for model, param_grid in zip(base_models, param_grids):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_estimators.append(grid_search.best_estimator_)
    print("Best parameters for {}: {}".format(model.__class__.__name__, grid_search.best_params_))
    print("Best MSE score for {}: {:.4f}".format(model.__class__.__name__, -grid_search.best_score_))
    print()

# Print best estimators
for model, model_string in zip(best_estimators, base_model_string):
    print("Best estimator for {}: {}".format(model_string, model))
    print()
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
import numpy as np

# Convert X to a DataFrame if it's not already
X = pd.DataFrame(X)

# Reset index of X to integer-based positional indexing
X.reset_index(drop=True, inplace=True)

# Define k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics for tuned models
MAE_tuned = []
MSE_tuned = []
RMSE_tuned = []
R2_tuned = []
EVS_tuned = []

# Train tuned models and perform k-fold cross-validation
i = 0
for model in best_estimators:
    mae_scores = []
    mse_scores = []
    rmse_scores = []
    r2_scores = []
    evs_scores = []
    
    for train_index, test_index in kf.split(X):
        X_train_cv, X_test_cv = X.iloc[train_index], X.iloc[test_index]
        y_train_cv, y_test_cv = y[train_index], y[test_index]
        
        model.fit(X_train_cv, y_train_cv)
        y_pred_cv = model.predict(X_test_cv)
        
        mae_scores.append(mean_absolute_error(y_test_cv, y_pred_cv))
        mse_scores.append(mean_squared_error(y_test_cv, y_pred_cv))
        rmse_scores.append(np.sqrt(mean_squared_error(y_test_cv, y_pred_cv)))
        r2_scores.append(r2_score(y_test_cv, y_pred_cv))
        evs_scores.append(explained_variance_score(y_test_cv, y_pred_cv))
    
    MAE_tuned.append(np.mean(mae_scores))
    MSE_tuned.append(np.mean(mse_scores))
    RMSE_tuned.append(np.mean(rmse_scores))
    R2_tuned.append(np.mean(r2_scores))
    EVS_tuned.append(np.mean(evs_scores))
    
    print("Performance evaluation of {} model".format(base_model_string[i]))
    print("Average MAE:", MAE_tuned[i])
    print("Average MSE:", MSE_tuned[i])
    print("Average RMSE:", RMSE_tuned[i])
    print("Average R^2 score:", R2_tuned[i])
    print("Average Explained Variance:", EVS_tuned[i])
    print()
    
    i += 1
```

```python
# Model Comparison of Baseline Models
df_tuned=pd.DataFrame({
    'Tuned ML Model':base_model_string,
    'MAE': MAE_tuned,
    'MSE': MSE_tuned,
    'RMSE': RMSE_tuned,
    'r^2 score':R2_tuned,
    'Explained Variance Score':EVS_tuned
})
df_tuned
```
